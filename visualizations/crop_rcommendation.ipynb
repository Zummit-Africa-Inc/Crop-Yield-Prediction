# Exploratory Data Analytics and Visualisation for Crop Recommendation
- The purpose of this script is to conduct Exploratory Data Analysis and Visualizations on the crop recommendation dataset, in order to uncover insights that could be useful for recommending crops tp plant.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read file into dataframe
df = pd.read_csv('Crop_recommendation.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.groupby(['label'])['ph'].agg(['mean', 'sum', 'count']).sort_values('sum', ascending=False).reset_index()
```

```python
df.groupby(['label'])['rainfall'].agg(['mean', 'sum', 'count']).sort_values('sum', ascending=False).reset_index()
```

```python
df.groupby(['label'])['humidity'].agg(['mean', 'sum', 'count']).sort_values('sum', ascending=False).reset_index()
```

```python
df.groupby(['label'])['temperature'].agg(['mean', 'sum', 'count']).sort_values('sum', ascending=False).reset_index()
```

```python
corr = df.select_dtypes(include = np.number).corr()

plt.figure(figsize = (8,8))
sns.heatmap(corr, annot = True)
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(x = 'temperature', y = 'humidity', hue ='label', data = df)
```

This shows that chickpeas, and kidneybeans are best produced at lower temperaturea and humidity. While Orange and Papaya are best produced at higher humidity and average to room temperature.
